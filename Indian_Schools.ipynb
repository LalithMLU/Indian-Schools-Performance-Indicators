1. What is the average number of schools per state?

3. What is the relationship between the number of schools and literacy rate?

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='Total_Schools', y='Literacy Rate (%)', data=dataset)
plt.title('Relationship between Total Schools and Literacy Rate')
plt.show()
```

4. What is the average pupil-teacher ratio across all states?

```python
average_pupil_teacher_ratio = dataset['Pupil_Teacher_Ratio'].mean()
average_pupil_teacher_ratio
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
sns.barplot(dataset,x='State',y='Pupil_Teacher_Ratio',palette='flare')
plt.xticks(rotation=45)
plt.xlabel('States')
plt.ylabel('Pupil Teacher Ratio')
plt.title('Pupil Teacher Ratio of Each Indian State')
plt.show()
```
